```python
from sklearn import svm
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve
import matplotlib.pyplot as plt
import pickle
import numpy as np
import seaborn as sns
import pandas as pd
from plotting_tools import plot_cm
```

```python
X_train = pd.read_pickle('training/x_train.pkl')
Y_train = pd.read_pickle('training/y_train.pkl')
X_test = pd.read_pickle('training/x_test.pkl')
Y_test = pd.read_pickle('training/y_test.pkl')
```

```python
X_train.shape
```

```python
X_train_5 = X_train[X_train['user'] < 5]
Y_train_5 = X_train_5.pop('user')
X_test_5 = X_test[X_test['user'] < 5]
Y_test_5 = X_test_5.pop('user')
```

```python
multiclass_5 = svm.LinearSVC()
multiclass_5.fit(X_train_5, Y_train_5)
```

```python
accuracy_score(multiclass_5.predict(X_test_5), Y_test_5)
```

```python
X_train_10 = X_train[X_train['user'] < 10]
Y_train_10 = X_train_10.pop('user')
X_test_10 = X_test[X_test['user'] < 10]
Y_test_10 = X_test_10.pop('user')
```

```python
X_train_10_to_19 = X_train[(X_train['user'] < 20) & (X_train['user'] > 9)]
Y_train_10_to_19 = X_train_10_to_19.pop('user')
X_test_10_to_19 = X_test[(X_test['user'] < 20) & (X_test['user'] > 9)]
Y_test_10_to_19 = X_test_10_to_19.pop('user')
```

```python
multiclass_10_to_19 = svm.LinearSVC()
multiclass_10_to_19.fit(X_train_10_to_19, Y_train_10_to_19)
```

```python
accuracy_score(multiclass_10_to_19.predict(X_test_10_to_19), Y_test_10_to_19)
```

```python
multiclass_10 = svm.LinearSVC()
multiclass_10.fit(X_train_10, Y_train_10)
```

```python
accuracy_score(multiclass_10.predict(X_test_10), Y_test_10)
```
