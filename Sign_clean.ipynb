# RDP
maximum size of sign is 1000 and the index of the sign is <h3><b>2243</b></h3> <br>
Applying RDP on the sign with 1000 data points to get the minimum points required for each sign


```python
#imports
import rdp
from rdp import rdp
```

```python
#i = 9
#plt.plot(df.X[i],df.Y[i])
def rdpp(x,y,e=1.3):
    l =zip(x,y)
    l=np.array(list(l))
    #change epsilon for more/less points.
    sc = rdp(l,epsilon = e)

    plt.plot(sc[:,0],sc[:,1])
    
    plt.plot(x,y)
    
    plt.show()
    print(len(sc))
rdpp(df.X[2243],df.Y[2243])
```

# Equidistant
100 points.

```python
from scipy.interpolate import interp1d
```

```python
# Linear length on the line
new_x = []
new_y = []
for i in range(4768):
    x = df.X[i]
    y = df.Y[i]
    distance = np.cumsum(np.sqrt( np.ediff1d(x, to_begin=0)**2 + np.ediff1d(y, to_begin=0)**2 ))

    distance = distance/distance[-1]
    #print(distance)
    fx, fy = interp1d( distance, x ), interp1d( distance, y )

    alpha = np.linspace(0, 1,100)
    x_regular, y_regular = fx(alpha), fy(alpha)
    new_x.append(x_regular)
    new_y.append(y_regular)
    print('done',i)


#plt.plot(x, y )
#plt.plot(x_regular, y_regular)
#plt.axis('equal')
#plt.show()
```

```python
df['new_x']=new_x
df['new_y']=new_y
```
